Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes.

```python
query_bike_id = """
SELECT bike_id, sum(duration_minutes) as duration
FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
GROUP BY bike_id
HAVING sum(duration_minutes) > 1000
ORDER BY duration DESC
"""
```

```python
bike_id = client.query(query_bike_id).to_dataframe()
bike_id
```

Filter trips where the trip_id is "Walk Up" and the end_station_name is "Stolen".

```python
query_filter_trip = """
SELECT trip_id
FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
WHERE subscriber_type LIKE '%Walk Up%' AND end_station_name LIKE '%Stolen%';
"""
```

```python
filter_trip = client.query(query_filter_trip).to_dataframe()
filter_trip
```
