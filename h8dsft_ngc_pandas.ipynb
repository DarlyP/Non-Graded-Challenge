Import Data

```python
# Import Library

import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/FTDS-learning-materials/phase-0/main/src/data_ngc.csv"
df = pd.read_csv(url, na_values=['N/A', 'NA'])
```

Pengecekan Missing Value

```python
# Melihat Jumlah Missing Value
df.isnull().sum()
```

```python
print('Total missing values in data:',df.isnull().sum().sum())
```

Catatan:

Terdapat banyak nilai missing value. Oleh karena itu pada kolom yang memiliki nilai value kurang dari 9000 akan dihapus. Kemudian akan dilakukan pengisian pada data yang memiliki nilai missing value. Untuk nilai yang numerical akan diberikan pengisian dengan interpolasi

```python
nilai_kosong = df.isnull().sum()
df_1 = df.dropna(axis=1, thresh=len(df) - 10000) # df_1 adalah dataframe yang tidak memiliki missing value
df_1.head()
```

```python
f'Bentuk data (row, column): {df_1.shape}'
```

```python
df_1_interpolated = df_1.interpolate()
df_1_interpolated.head()
```

```python
df_replaced = df_1_interpolated.replace(to_replace=[None], value='Unknown')
df_replaced.head()
```

```python
df_replaced.tail()
```

```python
df_replaced.info()
```

Catatan: Pada data di atas terlihat bahwa date masih berbentuk object sehingga perlu diubah menjadi bentuk datetime

```python
df_replaced['date'] = pd.to_datetime(df_replaced['date'])
```

```python
df_replaced.info()
```

Melakukan manipulasi tanggal

```python
df_replaced['Year'] = df_replaced['date'].dt.year
df_replaced['Month'] = df_replaced['date'].dt.month
df_replaced['Day'] = df_replaced['date'].dt.day
df_replaced
```

Membuat group data dari dataset

```python
# Mencari banyaknya yang meninggal di tahun 2022
f"Banyaknya yang meninggal di tahun 2022: {df_replaced[df_replaced['Year'] == 2022].count()['deaths']}"
```
